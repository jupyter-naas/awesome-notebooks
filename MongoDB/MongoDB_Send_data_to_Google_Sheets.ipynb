## Model

### Get data in MongoDB

```python
df = mongo.connect(host, port, user, passwd).get(collection_name, db_name)
```

## Output

### Send data to Google Sheet

```python
gsheet.connect(spreadsheet_id).send(
    sheet_name=sheet_name,
    data=df,
    append=False
)
```
