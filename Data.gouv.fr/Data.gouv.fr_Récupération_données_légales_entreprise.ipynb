## Input

### Import libraries

```python
import requests
from pprint import pprint
```

### Setup Variables

```python
# Input
SIREN = "877941724"
```

## Model

### Récupération des données

```python
def get_info(siren):
    req_url = f"https://entreprise.data.gouv.fr/api/sirene/v3/unites_legales/{siren}"
    res = requests.get(req_url)
    res.raise_for_status
    return res.json()
    
data = get_info(SIREN)
```

## Output

### Display result

```python
pprint(data)
```
