The objective of this notebook is to convert a CSV file into an Excel file.

## Input

### Import pandas

```python
import pandas as pd
```

### Variables

```python
excel_output = 'name-of-excel-file.xlsx'
```

## Models
Use pandas to: 
- read the CSV file link, 
- then convert it to Excel spreadsheet file by using the `to_excel` function. 

The sept helps to seperate the rows. Also, ensure the CSV link ends with a CSV file extension.

```python
convert_csv = pd.read_csv("input-csv-link-here", sep=",")
```

## Output

### Save result in .xlsx file

The header=True ensures the header from CSV file is converted.

```python
convert_csv.to_excel(excel_output, index=False, header=True) 
```
