## Input

### Import library

```python
from naas_drivers import mongo
import pandas as pd
```

### Variables

```python
# Credentials
user = "my user"
passwd = "my passwd"
host = "url"
port = 9090

# DB parameters
collection_name = "col"
db_name = "db_name"
```

## Model

### Read dataframe

```python
data = {"LABEL": "Label 1", "VALUE": 0}
df = pd.DataFrame([data])
df
```

## Output

### Send data

```python
mongo.connect(host, port, user, passwd).send(df, collection_name, db_name, replace=False)
```
