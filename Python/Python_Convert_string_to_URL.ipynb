## Input

### Import libraries

```python
import urllib.parse
```

### Setup Variables
- `string`: string to be converted to URL

```python
string = "This is a string"
```

## Model

### Convert string to URL

This function will convert a string to a URL using urllib.

```python
url = urllib.parse.quote(string)
```

## Output

### Display result

```python
print(url)
```
