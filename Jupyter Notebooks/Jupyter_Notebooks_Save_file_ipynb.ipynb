## Input

### Import libraries

```python
import json
from pprint import pprint
```

### Variables

```python
# Input
notebook_path = "../template.ipynb"

# Output
notebook_output = "new_template.ipynb"
```

## Model

### Open file

```python
with open(notebook_path) as f:
    nb = json.load(f)
```

### Save file

```python
with open(notebook_output, 'w') as f:
    json.dump(nb, f)
```

## Output

### Display result

```python
pprint(nb)
```
