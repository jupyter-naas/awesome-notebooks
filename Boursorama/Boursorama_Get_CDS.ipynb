## Input

### Import library

```python
import pandas as pd
```

### Variables

```python
URL = "https://www.boursorama.com/bourse/taux/cds/"
```

## Model

### Get the CDS from URL 

```python
dfs = pd.read_html(URL)
df = dfs[0]
```

## Output

### Display result

```python
df
```

### Save as csv

```python
df.to_excel(r'Get_CDS.csv', index = False)
```
