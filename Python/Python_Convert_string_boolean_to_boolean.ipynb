## Input

### Import libraries

```python
# Import libraries
import ast
```

### Setup Variables
- `string_boolean`: string boolean to convert

```python
# Setup variables
string_boolean = "True"
```

## Model

### Convert string boolean to boolean

```python
# Convert string boolean to boolean
boolean = ast.literal_eval(string_boolean)
```

## Output

### Display result

```python
# Display result
print(boolean)
```
