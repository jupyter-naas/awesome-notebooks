## Input

### Import libraries

```python
import pandas as pd
```

### Setup Variables
- `file_path`: Excel file path
- `sheet_name`: Strings are used for sheet names.

```python
file_path = ""
sheet_name = ""
```

## Model

### Read Excel

```python
df = pd.read_excel(file_path, sheet_name=sheet_name)
df
```

## Output

### Display result

```python
print("Row fetched:", len(df))
df.head(1)
```
