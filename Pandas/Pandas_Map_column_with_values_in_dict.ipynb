## Input

### Import libraries

```python
import pandas as pd
```

### Setup Variables

```python
data = {"Name": ["John", "Paul", "George", "Ringo"], "Age": [20, 21, 19, 18]}
mapping_dict = {20: "Young", 21: "Young", 19: "Adult", 18: "Adult"}
```

## Model

### Read dataframe

```python
df = pd.DataFrame(data)
df
```

### Map column with values in dict

```python
df["Age_Group"] = df["Age"].map(mapping_dict)
```
