### Setup MongoDB

### Setup Google Sheet
- Share your Google Sheet with our service account : ðŸ”— naas-share@naas-gsheets.iam.gserviceaccount.com

```python
spreadsheet_id = "------"
sheet_name = "Sheet1"
```

### Setup Naas

## Model

### Get Data From Google Sheets

```python
gsheet_data=naas_drivers.gsheet.connect(spreadsheet_id).get(sheet_name=sheet_name)
```

### Transform data to dict

```python
data = gsheet_data.to_dict("records")
```

## Output

### Send data to MongoDB

```python
naas_drivers.mongo.connect(host, port, user, passwd)
```

```python
data1 = pd.DataFrame(data)
naas_drivers.mongo.send(data1, collection_name, db_name,replace=False)
```
