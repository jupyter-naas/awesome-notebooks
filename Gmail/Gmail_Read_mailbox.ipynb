## Model

### Connect to email box

```python
emails = email.connect(username, 
        password, 
        username, 
        smtp_server)
```

### Get email list

```python
df = emails.get(criteria="unseen")
df
```

```python
uid_list = df['uid'].tolist() 
uid_list
```

### Setup the flags

```python
%%time
uid = uid_list
flag = "DELETED"
# possible value for flag:
# flag = 'SEEN'
# flag = 'ANSWERED'
# flag = 'FLAGGED'
# flag = 'DELETED'
# flag = 'DRAFT'
# flag = 'RECENT'
```
