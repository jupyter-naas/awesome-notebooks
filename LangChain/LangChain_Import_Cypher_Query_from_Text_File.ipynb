## Model

### Open and read a text file

```python
def read_txt_file(file_path):
    with open(file_path, 'r') as f:
        data = f.read()
    return data

import_query = read_txt_file(input_file_path)
import_query
```

## Output

### Import graph into Neo4j

```python
graph = Neo4jGraph(
    url=url, 
    username=username, 
    password=password
)
graph.query(import_query)
```
