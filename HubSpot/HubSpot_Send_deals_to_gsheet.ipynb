## Model

### Get deals

```python
df_deals = hubspot.connect(HS_ACCESS_TOKEN).deals.get_all(properties)
```

## Output

### Send contacts to gsheet

```python
gsheet.connect(SPREADSHEET_URL).send(sheet_name=SHEET_NAME, data=df_deals, append=False)
```
