## Input

### Import Libraries

```python
import requests
import json
```

### Setup Variables
- `api_key`: Your Trello API key. [Get your API key here](https://developer.atlassian.com/cloud/trello/guides/rest-api/authentication/)
- `token`: Your Trello token. [Get your token here](https://developer.atlassian.com/cloud/trello/guides/rest-api/authentication/)

```python
api_key = "your_api_key"
token = "your_token"
```

## Model

### Retrieve List of Boards

This function will retrieve a list of all the boards that you have access to in Trello.

```python
def get_boards(api_key, token):
    url = "https://api.trello.com/1/members/me/boards"
    querystring = {"key": api_key, "token": token}
    response = requests.request("GET", url, params=querystring)
    boards = json.loads(response.text)
    return boards
```

## Output

### Display Result

```python
boards = get_boards(api_key, token)
print(boards)
```
