### Import library

```python
from naas_drivers import airtable
```

### Variables

```python
API_KEY = 'API_KEY'
BASE_KEY = 'BASE_KEY'
TABLE_NAME = 'TABLE_NAME'
```

## Model

### Insert data

```python
data = airtable.connect(API_KEY,
                        BASE_KEY,
                        TABLE_NAME).get(view='All opportunities',
                                        maxRecords=20).insert({'Name': 'Brian'})
```
