## Input

### Import Libraries

```python
import requests
```

### Setup Variables
- `api_key`: Your Trello API key. [Get your API key here](https://trello.com/app-key).
- `token`: Your Trello token. [Get your token here](https://trello.com/app-key).
- `board_id`: The ID of the board you want to create the card in.
- `list_id`: The ID of the list you want to create the card in.
- `name`: The name of the card you want to create.
- `desc`: The description of the card you want to create.

```python
api_key = "YOUR_API_KEY"
token = "YOUR_TOKEN"
board_id = "BOARD_ID"
list_id = "LIST_ID"
name = "CARD_NAME"
desc = "CARD_DESCRIPTION"
```

## Model

### Create Card

This function will create a card on a Trello board using the API.

```python
def create_card(api_key, token, board_id, list_id, name, desc):
    url = "https://api.trello.com/1/cards"
    querystring = {
        "key": api_key,
        "token": token,
        "idBoard": board_id,
        "idList": list_id,
        "name": name,
        "desc": desc,
    }
    response = requests.request("POST", url, params=querystring)
    return response.json()
```

## Output

### Display Result

```python
card = create_card(api_key, token, board_id, list_id, name, desc)
print(card)
```
