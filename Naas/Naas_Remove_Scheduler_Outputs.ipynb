### Remove scheduler outputs.

```python
def remove_scheduler_outputs():
    for scheduler_output in glob.glob(
        "/home/ftp/.naas/home/ftp/**/*__output_*.ipynb", recursive=True
    ):
        os.remove(scheduler_output)
        print(f"âœ… DELETED: {scheduler_output}")
```
