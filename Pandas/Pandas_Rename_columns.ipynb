## Input

### Import libraries

```python
import pandas as pd
```

### Setup Variables
- `to_rename`: dict of columns with new names

```python
to_rename = {
    "team": "equipo",
    "points": "puntos",
    "assists": "asistancias",
    "rebounds": "rebotes",
    "col": "new_col",
}
```

## Model

### Create dataframe

```python
df = pd.DataFrame(
    {
        "team": ["A", "A", "A", "B", "B", "B"],
        "points": [11, 7, 8, 10, 21, 13],
        "assists": [5, 7, 7, 9, 12, 0],
        "rebounds": [5, 8, 10, 6, 6, 22],
    }
)
df
```

### Rename columns in dataframe

```python
df = df.rename(columns=to_rename)
```

### Display new DataFrame

```python
df
```
