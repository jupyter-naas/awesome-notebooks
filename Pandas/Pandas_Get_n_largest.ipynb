## Input

### Import libraries

```python
import pandas as pd
```

### Setup Variables

```python
# Create a fake dataset
data = {"Name": ["Tom", "nick", "krish", "jack"], "Age": [20, 21, 19, 18]}
# Create a DataFrame
df = pd.DataFrame(data)
```

## Model

### Get n largest

The `nlargest` function can be used to get the n largest values from a DataFrame.

```python
# Get the two largest values
df.nlargest(2, "Age")
```

## Output

### Display result

The two largest values are displayed below.
